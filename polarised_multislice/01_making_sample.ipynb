```python
import jax.numpy as jnp
import matplotlib.pyplot as plt
from sample import paper_sample, single_bead_sample

%config InlineBackend.figure_format='retina'
```

```python
# Settings from paper
n_m = 1.33
n_bead = jnp.array([1.44, 1.40, 1.37]) # z y x
bead_radius = 1.5
spacing = 0.065 # mum
k0 = 2 * jnp.pi / 0.405 #mum
```

```python
# Making a single bead
orientation = jnp.array([0, 0, jnp.pi/2]) # z y x 
shape = (4.55, 11.7 / 2, 11.7 / 2) # z y x # / 2 as we have only 1 bead

sample = single_bead_sample(n_m, n_bead, orientation, bead_radius, shape, spacing, k0)
print(f"Shape of sample {sample.shape}")
```

```python
plt.title("Cross section of sample")
plt.imshow(sample[35, :, :, 2, 2])
plt.colorbar()
```

```python
print("Scattering potential tensor inside bead")
with jnp.printoptions(precision=2, suppress=True):
    print(sample[35, 45, 45])
```
